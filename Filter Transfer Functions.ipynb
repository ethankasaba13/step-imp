# Filter Transfer Functions

## Imports

```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt
```

## Variables

```python
ord = 4 # order, number of reactive elements in the filter
rip = 0.1 # maximum ripple (dB) in pass band for Chebyshev I and elliptical filters
minatt = 10 # minimum attenuation (dB) required in the stop band for Chebyshev II and elliptical filters
cutoff = 2.5 # cutoff frequency (GHz)
type = 'low' # type of filter: low-pass, high-pass, etc.

# Variable manipulation
cutoff *= np.pi * 20**9 # conversion of cutoff freq to rad/s for use in scipy
```

## Chebyshev Type 1 Filter

```python
b, a = signal.cheby1(ord, rip, cutoff, type, analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w / (2 * np.pi), 20 * np.log10(abs(h)))
plt.title('Chebyshev Type I frequency response')
plt.xlabel('Frequency [Hertz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(cutoff / (2 * np.pi), color='green') # cutoff frequency
plt.show()
```

## Chebyshev Type 2 Filter

```python
b, a = signal.cheby2(ord, minatt, cutoff, type, analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w / (2 * np.pi), 20 * np.log10(abs(h)))
plt.title('Chebyshev Type II frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(cutoff / (2 * np.pi), color='green') # cutoff frequency
plt.show()
```

## Butterworth Filter

```python
b, a = signal.butter(ord, cutoff, type, analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w / (2 * np.pi), 20 * np.log10(abs(h)))
plt.title('Butterworth frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(cutoff / (2 * np.pi), color='green') # cutoff frequency line
plt.show()
```

## Elliptical Filter

```python
b, a = signal.ellip(ord, rip, minatt, cutoff, type, analog=True)
w, h = signal.freqs(b, a)
plt.semilogx(w / (2 * np.pi), 20 * np.log10(abs(h)))
plt.title('Elliptical frequency response')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1)
plt.grid(which='both', axis='both')
plt.axvline(cutoff / (2 * np.pi), color='green') # cutoff frequency line
plt.show()
```
